# Traverser Function

```python
def get_report_df(report_list):
    
    # Save an empty mega report card dataframe
    mega_report_df=pd.DataFrame()
    
    #######################################################################################  
    # For Loop
    # For each class dictionary in the list:
    for division in report_list:
    
        # Make Scores Df
        scores_df=pd.DataFrame(division["scores"])
    
    
        # Make Emails Df
        emails_df = pd.DataFrame(division["emails"])
    
    
        # Merge the Scores df with the Emails Df
        scores_emails_df = pd.merge(scores_df,emails_df, how= "left" , on=["student_name"])
    
    
        # Add Class column
        scores_emails_df["class"]= division["class"]
    
        # Add Division column
        scores_emails_df["division"]= division["division"]
    
    
        # Add Teacher column
        scores_emails_df["teacher"]=division["teacher"]
    
        # Add (Concat) it to a mega report card dataframe
        mega_report_df=pd.concat([mega_report_df,scores_emails_df], ignore_index = True)
    ###########################################################################################
    # Return mega report card dataframe
    
       
    mega_report_df["total_scores"]=mega_report_df["math_scores"]+mega_report_df["chemistry_scores"]+mega_report_df["information_technology_scores"]+mega_report_df["physics_scores"]
    mega_report_df["average_score"]=mega_report_df["total_scores"]/4
    
    
    mega_report_df.loc[mega_report_df["average_score"]>=35, "status"]= "pass"
    mega_report_df.loc[mega_report_df["average_score"]<35, "status"]= "fail"
    
    
    mega_report_df.loc[mega_report_df["average_score"]>=90 , "grades"]= "A"
    mega_report_df.loc[(mega_report_df["average_score"]>=75) & (mega_report_df["average_score"]<90), "grades"]="B" 
    mega_report_df.loc[(mega_report_df["average_score"]>=50) & (mega_report_df["average_score"]<75), "grades"]="C" 
    mega_report_df.loc[(mega_report_df["average_score"]>=35) & (mega_report_df["average_score"]<50), "grades"]="D" 
    mega_report_df.loc[mega_report_df["average_score"]<35 , "grades"]= "F"

    
    
    return mega_report_df
```

# How many students passed and how many students failed in each division?
